```python
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

auth.authenticate_user()
project_id = "gold-yen-408708" # Ini project ID
client = bigquery.Client(project=project_id)
```

```python
query = """SELECT
	A.YEAR,
	A.MONTH,
	COUNT(A.MONTH) AS TOTAL_PATENT
FROM
	(SELECT
		EXTRACT(YEAR FROM (PARSE_DATE('%Y%m%d', Filing_Date))) AS YEAR,
		EXTRACT(MONTH FROM (PARSE_DATE('%Y%m%d', Filing_Date))) AS MONTH
	FROM patents-public-data.uspto_oce_cancer.publications
	WHERE
		EXTRACT(YEAR FROM (PARSE_DATE('%Y%m%d', Filing_Date))) BETWEEN 2012 AND 2023) AS A
GROUP BY
	A.YEAR, A.MONTH
ORDER BY
	A.YEAR DESC,
	A.MONTH ASC;"""
# Set up the query

query_job = client.query(query).to_dataframe()
query_job
```
