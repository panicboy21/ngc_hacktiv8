```python
query = """SELECT count(*) as TOTAL_TABLE FROM bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.TABLES;"""
# Set up the query

query_job = client.query(query).to_dataframe()
query_job
```

Total Column pada Table `bikeshare_stations` yang mempunyai data type STRING

```python
query = """SELECT count(*) as Jumlah_Column_String FROM bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS WHERE table_name = 'bikeshare_stations' AND data_type = 'STRING';"""
# Set up the query

query_job = client.query(query).to_dataframe()
query_job
```

Retrieve all trips where the bike_id is not null and the duration is greater than 1000 minutes

```python
query = """SELECT * FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` WHERE bike_id IS NOT NULL AND duration_minutes > 1000;
 """
# Set up the query

query_job = client.query(query).to_dataframe()
query_job
```

Filter trips where the trip_id is "Walk Up" and the end_station_name  is "Stolen"

```python
query = """SELECT * FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` WHERE subscriber_type = 'Walk Up' AND end_station_name ='Stolen';
 """
# Set up the query

query_job = client.query(query).to_dataframe()
query_job
```
