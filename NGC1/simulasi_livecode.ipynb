Buat program untuk stock management yang dapat melakukan:
1. User dapat menginput barang, kategori, harga barang dan tersimpan dalam nested list
2. User dapat menampilkan list stock barang yang sudah diinput, jika belum ada barang yang dinput di stock maka saat pilih menampilkan barang akan muncul pesan "Data Belum Ada"





```python
class stock_management():
    def __init__(self):
        self.list_barang = [] #varible penampung barang dalam nested list
    
    def addbarang (self, nama_barang, kategori_barang, harga_barang): #function add barang
        self.nama_barang = nama_barang
        self.kategori_barang = kategori_barang
        self.harga_barang = harga_barang

        list_add = [] # list temp untuk nested list barang
        list_add.append(self.nama_barang)
        list_add.append(self.kategori_barang)
        list_add.append(self.harga_barang)

        self.list_barang.append(list_add) #penambahan list ke dalam nested list
    
    def print_list_barang(self):
        # Cek data dalam list ada atau tidak
        if not any(self.list_barang):
            print("")
            print("Data Belum Ada")
        else:
            print ("")
            print ("NO.|Nama Barang|Kategori Barang|Harga")
            for index, i in enumerate(self.list_barang): # print data sesuai nested list data
                nomor = index+1
                print(nomor, *i, sep=" | ")

def mainfunction():
    try:
        objStock = stock_management()
        print("============================")
        print("Program Inventory Management")
        print("============================")
        
        while True:
            print("")
            print("Menu :")
            print("1. Input Data")
            print("2. Menampilkan Data")
            pilih = input("Pilih Menu atau Keluar : ")
            
            # Cek pilihan yang dipilih user
            if pilih == "1":
                nama_barang = input("Masukan Nama Barang : ")
                kategori_barang = input("Masukan Kategori Barang : ")
                harga_barang = int(input("Masukan Harga Barang : "))
                objStock.addbarang(nama_barang, kategori_barang, harga_barang)
                print("")
                print("Terimakasih, data sudah disimpan")
            elif pilih == "2":
                objStock.print_list_barang()
            elif pilih == "exit":
                print("")
                print("Terima Kasih, Sampai Jumpa!")
                break
            else:
                print("")
                print("Inputan menu salah, silahkan ulangi")

    except:
        print("Ada Error")

if __name__ == "__main__":
    mainfunction()
```
